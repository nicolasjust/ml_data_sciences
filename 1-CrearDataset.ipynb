# Creacion del dataset
En esta seccion se realizan las diferentes llamados apis de ML y se crea un unico dataset

```python
import pandas as pd
import requests
import seaborn as sns
import math
from time import time
```

```python
import requests_cache
requests_cache.install_cache() # Para cachear algunos los llamados a la API
```

```python
import os 
# Saco un token de acceso para poder acceder a todas las paginas de resultados
token = os.environ.get('ML_TOKEN')
```

```python
keywords = [
    'tv 4k','tv','tv smart', 'celulares', 'galaxy','iphone','apple','notebook', 'computadora',
    'laptop','computadora gamer','acer','playstation','ps5','ps4','xbox','home theater', 'sony',
    'heladera','lavarropa','samsung','gafa','microondas','electrodomesticos','xiaomi','gamer','monitor',
    'dell','lg','computacion','camaras','electronica','audio','consolas','videojuegos','audio','impresoras',
    'tablet'
]
```

```python
# Como los resultados llamando a la api /search y /items tienen diferentes atributos junto todo en uno
resultados=[]
for k in keywords:
    # Calculo las paginas
    result = requests.get("https://api.mercadolibre.com/sites/MLA/search", params={'q':k,'limit':1}) # Para calcular cuantas paginas hay
    jsonResult = result.json()
    cantidad_paginas = math.ceil(min(jsonResult['paging']['total'] / 50, 70)) # Pongo un tope de 50 paginas por keyword
    #print("Para el termino %s hay %s paginas"%(k, cantidad_paginas))


    for page in range(cantidad_paginas):
        try:
            start_time = time()
            result = requests.get("https://api.mercadolibre.com/sites/MLA/search", params={'q':k,'offset':50*page}
                                 ,headers={'Authorization': 'Bearer  %s'%token}) # Para calcular cuantas paginas hay
            jsonResult = result.json()
            for resultSearch in jsonResult['results']:
                idProducto = resultSearch['id']
                idCategoria = resultSearch['category_id']
                if len([x for x in resultados if x['id']==idProducto]) == 0: # Solo incluyo productos nuevos, asi no hay duplicados
                    resultItem = requests.get("https://api.mercadolibre.com/items/"+idProducto).json()
                    resultCategory = requests.get("https://api.mercadolibre.com/categories/"+idCategoria).json()

                    for categ in resultCategory.copy().keys():
                        resultCategory['category_'+categ] = resultCategory.pop(categ)

                    resultFinal = {**resultItem, **resultSearch,**resultCategory} # Diccionario con todos los atributos
                    resultados.append(resultFinal)
            print("Se proceso la pagina  pagina %s/%s -- %.10f segundos"%(page, cantidad_paginas,time() - start_time))
        except e:
            print("Error")
```

```python
print("Cantidad de resultados",len(resultados))
```

```python
df = pd.json_normalize(resultados)
df.set_index('id',inplace=True)
df.head()
```

```python
# Incluyo tambien los atributos de los productos y terminos de ventas
list_atributos = []
for product_id,productAttibutes in df.iterrows():
    todosAtributos = {'id':product_id}
    for at in productAttibutes['attributes']:
        todosAtributos = {**todosAtributos,**{"product_"+at['name'] :at['value_name']}}

      
    for at in productAttibutes['sale_terms']:
        todosAtributos = {**todosAtributos,**{"sale_terms_"+at['name'] :at['value_name']}}

    list_atributos.append(todosAtributos)

df_atributos = pd.DataFrame(list_atributos)
df_atributos.set_index('id',inplace=True)
```

```python
# Dataframe completo
df = df.join(df_atributos)
```

```python
df.to_csv('dataset.csv')
```
